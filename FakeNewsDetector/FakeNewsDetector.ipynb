```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
fakeData = pd.read_csv('fake.csv', engine='python', encoding='utf-8', error_bad_lines=False)
trueData = pd.read_csv('true.csv', engine='python', encoding='utf-8', error_bad_lines=False)
```

```python
data_merge = pd.concat([fakeData, trueData], axis = 0)
data_merge.head(10)
```

```python
data = data_merge.drop(['title', 'subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data
```

```python
def wordOpt(text):
  text = text.lower()
  text = re.sub('\[.*?]', '', text)
  text = re.sub('\\W', ' ', text)
  text = re.sub('https?://\S+\ww\.S+', '', text)
  text = re.sub('<.*?>+', '', text)
  text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
  text = re.sub('\n', '', text)
  text = re.sub('\w*\d\w*', '', text)
  return text
```

```python
data['text'] = data['text'].apply(wordOpt)
```

```python
x = data['text']
y = data['class']
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x,y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
x_train = vectorization.fit_transform(xTrain)
x_test = vectorization.transform(xTest)
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(x_train, yTrain)
```

```python
pred_LR = LR.predict(x_test)
```

```python
LR.score(x_test, yTest)
```

```python
#LETS GOOOOOOOOO HIGH ACCURACY WOOOOOOOOOOOOOOOOOOOOOOO
```

```python
print(classification_report(yTest, pred_LR))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(x_train, yTrain)
```

```python
pred_DT = DT.predict(x_test)
```

```python
DT.score(x_test, yTest)
```

```python
print(classification_report(yTest, pred_DT))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0)
GB.fit(x_train, yTrain)
```

```python
pred_GB = GB.predict(x_test)
```

```python
GB.score(x_test, yTest)
```

```python
print(classification_report(yTest, pred_GB))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier()
RF.fit(x_train, yTrain)
```

```python
pred_RF = RF.predict(x_test)
```

```python
RF.score(x_test, yTest)
```

```python
print(classification_report(yTest, pred_RF))
```

```python
def output_lable(n):
  if n == 0:
    return "Fake News"
  elif n == 1:
    return "Not A Fake News"
def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(wordOpt)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  pred_LR = LR.predict(new_xv_test)
  pred_DT = DT.predict(new_xv_test)
  pred_GB = GB.predict(new_xv_test)
  pred_RF = RF.predict(new_xv_test)

  return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_LR[0]),
                                                                                                            output_lable(pred_GB[0]),
                                                                                                            output_lable(pred_RF[0]),
                                                                                                            output_lable(pred_DT[0])))
```
